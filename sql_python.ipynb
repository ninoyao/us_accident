```python
import pymysql
import seaborn as sns
import pandas as pd
sns.set

#conn = pymysql.connect('localhost','root','root','accidents2',port=8889)
```

```python
def fetch_clean1(localhostname,user,pw,db,port,table):
    conn = pymysql.connect(localhostname,user,pw,db,port)
    cursor = conn.cursor()

    #accidents data
    cursor.execute(f'SELECT * FROM {table}')
    desc = cursor.description
    data = cursor.fetchall()

    data_names = []
    for i in range(len(desc)):
        data_names.append(desc[i][0])

    data_df = pd.DataFrame(data, columns = data_names)
    conn.close()
    return data_df
```
